### 8. Compare the number of stops during early morning (0–6 AM) vs night (6 PM–12 AM).

```python
data['stop_time'] = pd.to_datetime(data['stop_time'], format='%H:%M', errors='coerce')
data['hour'] = data['stop_time'].dt.hour

data[(data.hour >= 0) & (data.hour < 6)].shape[0], data[(data.hour >= 18) & (data.hour <= 23)].shape[0]
```

```python
data['stop_time'] = pd.to_datetime(data['stop_time'], format='%H:%M', errors='coerce')
data['hour'] = data['stop_time'].dt.hour
early_morning_stops = data[(data['hour'] >= 0) & (data['hour'] < 6)].shape[0]
night_stops = data[(data['hour'] >= 18) & (data['hour'] < 24)].shape[0]
print(f"Number of stops during early morning (0-6 AM): {early_morning_stops}")
print(f"Number of stops during night (6 PM-12 AM): {night_stops}")
```
